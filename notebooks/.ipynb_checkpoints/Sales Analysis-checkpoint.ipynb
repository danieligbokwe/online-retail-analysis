Project 1: Online Retail Sales Analysis
This notebook analyzes the online retail dataset to uncover trends in products, revenue, and countries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv('../data/online_retail.csv', encoding='ISO-8859-1')
```

```python
# Remove cancelled orders
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# How many rows remain
print("Rows after removing cancellations:", df.shape[0])
```

```python
# Check missing values in key columns
print(df[['CustomerID', 'Description']].isnull().sum())
```

```python
# Drop rows with missing CustomerID or Description
df = df.dropna(subset=['CustomerID', 'Description'])

# Confirm the change
print("Rows after dropping missing values:", df.shape[0])
```

```python
# Remove exact duplicate rows
df = df.drop_duplicates()

# How many rows remain
print("Rows after removing duplicates:", df.shape[0])
```

```python
# Ensure InvoiceDate is a datetime object
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create Revenue column
df['Revenue'] = df['Quantity'] * df['UnitPrice']

# Extract Year and Month for trend analysis
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month

# Preview the updated dataset
df.head()
```

```python
# Keep only positive quantities (actual sales)
df = df[df['Quantity'] > 0]

# Check the number of rows remaining
print("Rows after removing negative quantities:", df.shape[0])
```

```python
# Top 10 products by total revenue
top_products = df.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)

# Display the result
print(top_products)

# Plot the top 10 products
plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values, y=top_products.index, palette="viridis")
plt.title('Top 10 Products by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Product')
plt.show()
```

```python
# Total revenue by country
country_revenue = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(5)

# Display the result
print(country_revenue)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=country_revenue.values, y=country_revenue.index, palette="magma")
plt.title('Top 5 Countries by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Country')
plt.show()
```

```python
# Monthly revenue trend
monthly_revenue = df.groupby(['Year','Month'])['Revenue'].sum().reset_index()
monthly_revenue['Date'] = pd.to_datetime(monthly_revenue[['Year', 'Month']].assign(DAY=1))

# Plot the trend
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_revenue, x='Date', y='Revenue', marker='o')
plt.title('Monthly Revenue Trend For 2011')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total revenue per customer
top_customers = df.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False).head(10)

# Display
print(top_customers)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=top_customers.values, y=top_customers.index.astype(str), palette="coolwarm")
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Revenue')
plt.ylabel('CustomerID')
plt.show()
```

```python
# Average monthly revenue across all years
monthly_avg = df.groupby('Month')['Revenue'].mean()

# Plot
plt.figure(figsize=(12,6))
sns.lineplot(x=monthly_avg.index, y=monthly_avg.values, marker='o', color='green')
plt.title('Average Monthly Revenue Over The Years')
plt.xlabel('Month')
plt.ylabel('Average Revenue')
plt.xticks(range(1,13))
plt.show()
```

```python
import sqlite3
conn = sqlite3.connect(':memory:')
df.to_sql('online_retail', conn, index=False, if_exists='replace')
cur = conn.cursor()
```

```python
query = """
SELECT Description, SUM(Revenue) AS TotalRevenue
FROM online_retail
GROUP BY Description
ORDER BY TotalRevenue DESC
LIMIT 10;
"""

top_products_sql = pd.read_sql(query, conn)
print(top_products_sql)
```

```python
query = """
SELECT Country, SUM(Revenue) AS TotalRevenue
FROM online_retail
GROUP BY Country
ORDER BY TotalRevenue DESC
LIMIT 5;
"""

top_countries_sql = pd.read_sql(query, conn)
print(top_countries_sql)
```

```python
query = """
SELECT strftime('%Y-%m', InvoiceDate) AS YearMonth, SUM(Revenue) AS TotalRevenue
FROM online_retail
GROUP BY YearMonth
ORDER BY YearMonth;
"""

monthly_trend_sql = pd.read_sql(query, conn)
monthly_trend_sql
```
